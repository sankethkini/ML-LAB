```python
import random
def clustering(mylist,numofcen,it):
    centers=[]
    dim=len(mylist[0])
    for i in range(numofcen):
        for j in range(dim):
            if i>len(centers)-1:
                centers.append([random.uniform(1,3)])
            else:
                centers[i].append(random.uniform(1,5))
    clusters=[0]*len(mylist)
    for _ in range(it):
        print(centers)
        for i in range(len(mylist)):
            close=findclosest(mylist[i],centers)
            clusters[i]=close
        centers=update(clusters,mylist,centers)
    return clusters


def update(clusters,mylist,centers):
    mydict={}
    for i in range(len(mylist)):
        if mydict.get(clusters[i])==None:
            mydict[clusters[i]]=[mylist[i]]
        else:
            mydict[clusters[i]].append(mylist[i])
    dim=len(mylist[0])
    point=[0]*dim
    for key,val in mydict.items():
        for i in range(dim):
            d=0
            for j in range(len(val)):
                d+=val[j][i]
            point[i]=d/len(val)
        centers[key]=point
    return centers

def findclosest(point,centers):
    dist=[]
    for ct in centers:
        d=0
        for c in range(len(ct)):
            d+=(ct[c]-point[c])**2
        d=d**0.5
        dist.append(d)
    retval=0
    minval=dist[0]
    for i in range(1,len(dist)):
        if dist[i]<minval:
            retval=i
            minval=dist[i]
    return retval
```

```python
mylist=[[100,200],[100,50],[50,60],[1200,800],[1000,500]]
clustering(mylist,3,40)
```
